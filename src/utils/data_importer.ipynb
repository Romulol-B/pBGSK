```python

from ucimlrepo import fetch_ucirepo 
DATASET_REGISTRY = {
"tic_tac_toe_endgame": 101, 
"breast_cancer":  15 ,
"wine_quality":  193 ,
"heart_disease":  53 ,
"house_votes":  29 ,
"zoo":  30 ,
"Lymphography":  63 ,
"hepatitis":  37 ,
#medium size datasets
"waveform":  105 ,
"german_credit":  31 ,
"wdbc":  151 ,
"ionosphere":  146 ,
"dermatology":  45 ,
"soybean":  50 ,
"lung_cancer":  60 ,
"spambase":44 ,
"sonar":  147 ,
#large size datasets
"hill_valley":  106 ,
"clean1":  189 ,
"semeion":  49 ,
"arryhythmia":  148 ,
"cnae":  192}
```

```python
def data_loader (dataset_name):
    if dataset_name in DATASET_REGISTRY:
        dataset_id = DATASET_REGISTRY[dataset_name]
        dataset = fetch_ucirepo(id=dataset_id)
        X = dataset.data
        y = dataset.target
        return X, y
    else:
        raise ValueError(f"Dataset '{dataset_name}' not found in registry.")
```
